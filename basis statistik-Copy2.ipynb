```python

import pandas as pd

# the length for text
train = pd.read_csv('train.csv')
train['length'] = train['text'].str.len()

train.head()
```

```python
# the after padding length for train according to length
train['afterPadding'] = train['length'].map(lambda x: (x+100)//100*100)
train.groupby('afterPadding').count()
```

```python
# the length for ciphertext
test = pd.read_csv('test.csv')
test['length'] = test['ciphertext'].str.len()

test.head()
```

```python
# test1 is the table of all the ciphertext of difficulty 1
test1 = test.loc[test['difficulty'].isin([1])]

# the length for ciphertext of difficulty 1
test1['length'] = test1['ciphertext'].str.len()
test1.groupby('length').count()
```

```python
# find the 1. and the last space in plain text
train['1.space'] = train['text'].str.find(' ')
train['last space'] = train['text'].str.rfind(' ')


# find the length of padding at the beginning
train['padding'] = train['length'].map(lambda x: (((x+100)//100*100) - x)//2)

# find the number of spaces
train['space count'] = train['text'].str.count(' ') 

train.head()
```
